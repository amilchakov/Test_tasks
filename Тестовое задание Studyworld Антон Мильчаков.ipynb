```python
# удаление полных дублей
df = df.drop_duplicates()
```

```python
# удаление полей, где значение is_paid не определено, их невозможно интерпретировать
df = df.dropna(subset = ['is_paid'])
```

```python
# сортировка по всем колонкам в датафрейме
df = df.sort_values(df.columns.tolist())
```

```python
# удаление оставшихся дублей
df = df.drop_duplicates(subset = ['order_id'])
df = df.drop([1189])
```

```python
# создание колонки со сроком выполнения заказа в часах
df['term'] = ((df.deadline - df.creation)/np.timedelta64(1, 'h')).round(0)
```

```python
# сохранение очищенного датафрейма
df.to_csv('test_data_clear.csv')
```

Дальнейшие расчеты и построение визуализации были проведены в Tableau

Ссылка на аналитический дашборд - https://public.tableau.com/app/profile/anton.milchakov/viz/_16361388804460/Dashboard1

Были проверены гипотезы по влиянию следующих полей заказа на конверсию в оплату:

 - title - длина названия заказа
 - description - длина описания заказа
 - term (deadline - creation) - срок выполнения заказа
 - file_count - кол-во приклепленных файлов к заказу за первые 10 минут
 - wordsFrom - нижняя граница слов в выполненном заказе
 - wordsTo - верхняя граница слов в выполненном заказе

### Выводы:

##### Общее:
 - После очистки исходных данных осталось 9333 уникальных заказа с заполненным полем is_paid в период с 01.09.2015 по 27.10.2015
 - Из них оплачено - 872
 - Конверсия в оплату - 9.34%

##### Поле title:
 - заказы с очень коротким (длина до 10) и очень длинным (длина более 60) названием имеют конверсию ниже чем заказы с длиной от 11 до 60

##### Поле description:
 - прямая коррелляция конверсии и длины описания - чем длиннее описание тем выше конверсия
 - заказы с незаполненым полем имеют наименьший показатель

##### Поля deadline и creation:
 - была проведена оценка влияния срока выполнения заказа (term) на конверсию
 - низкие показатели у срочных заказов (до 8 часов) и заказов с большим сроком (более 6 суток)
 - наибольшая конверсия у заказов, которые необходимо выполнить в течение 3-6 суток
 
##### Полe file_count:
 - наличие прикрепленных файлов оказывает значительное влияние на конверсию (39.55% - есть прикрепленные файлы, 8.04 % - нет)
 
##### Поля wordsFrom и wordsTo:
 - значения в полях зависят от типа заказа, поэтому нет отчетливой тенденции.
 - возможно, стоит рассмотреть возможность удаления этих полей из формы заказа, т.к. об объеме работы заказчик может сообщить в описании или в переписке с исполнителем.

 
В дашборде есть фильтры, где можно посмотреть на показатели конверсии в различных разрезах:
 - type_id - тип заказа
 - category_id - предмет заказа
 - source_of_user - источник пользователя
 - is_test - тестовый ли пользователь
 
В целом, по различным разрезам тенденции сохраняются, за исключением позиций с небольшим количеством заказов

Ниже топы по количеству заказов в различных разрезах:

```python
df.groupby('type_id') \
  .agg({'order_id': 'count'}) \
  .sort_values('order_id', ascending=False) \
  .head()
```

```python
df.groupby('category_id') \
  .agg({'order_id': 'count'}) \
  .sort_values('order_id', ascending=False) \
  .head()
```

```python
df.groupby('source_of_user') \
  .agg({'order_id': 'count'}) \
  .sort_values('order_id', ascending=False)
```
